# Função de Ativação

```
import numpy as np


def funcao_ativacao(entrada_funcao):
  '''Função de ativação degrau para o neurônio'''
  if entrada_funcao >=0:
    return 1
  else:
    return 0

def neuronio(entrada, peso, bias):
  '''
  Função do neurônio:
  entrada: x
  peso: w
  bias: b
  '''

  #Chamar a função .dot() da biblioteca numpy para o cálculo do produto escalar
  u = np.dot(entrada, peso) + bias
  y = funcao_ativacao(u)
  return y
```

```
entrada = np.array([1,0])
pesos = np.array([0.6, 0.2])
bias = -0.5
saida = neuronio(entrada, pesos, bias)
print("Saída do neurônio: ", saida)
```

#Treinamento de Perceptron

```
import numpy as np
import matplotlib.pyplot as plt

class Perceptron:
  def __init__(self, n_input, taxa_aprendizado = 0.1) -> None:
    self.pesos = [0] * n_input
    self.bias = 0
    self.eta = taxa_aprendizado

  def ativacao(self, entrada):
    if entrada >= 0:
      return 1
    else:
      return 0

  def prever(self, entradaX):
    '''Saída da rede'''
    soma = 0
    for index in range(len(entradaX)):
      soma += entradaX[index]*self.pesos[index]
    soma = soma + self.bias
    y = self.ativacao(soma)
    return y

  def treinar(self, entrada, saida, epocas):
    for epoca in range(epocas):
      for count in range(len(entrada)):
        saida_rede = self.prever(entrada[count])
        erro = saida[count] - saida_rede

        for index in range(len(self.pesos)):
          self.pesos[index] += self.eta * erro * entrada[count][index]
        self.bias += self.eta * erro
      print("Epocas: ", epoca, "Pesos: ", self.pesos)
```

```
entradas = [(0,0), (0,1), (1,0), (1,1)]
saida_desejada = [0, 1, 1, 1]

rede = Perceptron(n_input = 2, taxa_aprendizado = 0.1)
rede.treinar(entradas, saida_desejada, 20)

for entrada in entradas:
  print("Entrada: ", entrada, "Saída prevista: ", rede.prever(entrada))
```
